```python
# X and Y column defining
X=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']]
y=df[["Chance of Admit "]]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
#train test splitting datas
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
#fitting Model
des=DecisionTreeRegressor()
des.fit(X_train,y_train)
```

```python
# Y prediction 
y_pred=des.predict(X_test)
```

```python
# Calculate accuracy and other metrics

from sklearn.metrics import mean_squared_error, r2_score
r2=r2_score(y_test, y_pred)
ms=mean_squared_error(y_test, y_pred)

print(f"Mean Square Error: {ms}\nR Sq error : {r2}")
```

## Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor(random_state=0)
regr.fit(X, y)
```

```python
y_pred1=regr.predict(X_test)
```

```python
# Calculate accuracy and other metrics

from sklearn.metrics import mean_squared_error, r2_score
r2_2=r2_score(y_test, y_pred1)
ms_2=mean_squared_error(y_test, y_pred1)

print(f"Mean Square Error: {ms_2}\nR Sq error : {r2_2}")
```
